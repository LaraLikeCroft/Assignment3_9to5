````sql
USE LaborStatisticsDB
SELECT s.series_id, s.series_title, i.industry_code, i.industry_name, ss.supersector_code, ss.supersector_name
FROM dbo.series s

JOIN dbo.industry i ON i.industry_code = s.industry_code
JOIN dbo.supersector ss ON ss.supersector_code = s.supersector_code

WHERE s.series_title = 'women employees' AND i.industry_name LIKE '%commercial banking%' AND ss.supersector_name LIKE '%financial activities%'
````

1. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Write a query that returns the series_id, industry_code, industry_name, and value from the january_2017 table but only if that value is greater than the average value for annual_2016 of data_type_code 82.</span>

````sql
USE LaborStatisticsDB
SELECT s.series_id, i.industry_code, i.industry_name, j.value
FROM dbo.january_2017 j

LEFT JOIN dbo.series s ON s.series_id = j.series_id
LEFT JOIN dbo.industry i ON i.industry_code = s.industry_code

WHERE j.value >
    (SELECT round(avg(a.value),2)
    FROM dbo.annual_2016 a

    LEFT JOIN dbo.series s ON a.series_id = s.series_id
    LEFT JOIN dbo.datatype d ON s.data_type_code = d.data_type_code
    WHERE d.data_type_code = 82)

ORDER BY j.value

/* Checking - 819960.34 */
/*SELECT round(avg(a.value),2)
    FROM dbo.annual_2016 a

    LEFT JOIN dbo.series s ON a.series_id = s.series_id
    LEFT JOIN dbo.datatype d ON s.data_type_code = d.data_type_code
    WHERE d.data_type_code = 82 */
````

2. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Create a Union table comparing average weekly earnings&nbsp; of production and nonsupervisory employees between annual_16 and january_17 using the data type 30.&nbsp; Round to the nearest penny.&nbsp; You should have a column for the average earnings and a column for the year, and the period.</span>

````sql
USE LaborStatisticsDB
SELECT round(avg(j.value),2) AS average_earnings, j.year, j.period
FROM dbo.january_2017 j

LEFT JOIN dbo.series s ON s.series_id = j.series_id
LEFT JOIN dbo.datatype d ON d.data_type_code = s.data_type_code

WHERE d.data_type_code = 30

GROUP BY j.year, j.period

UNION 

SELECT round(avg(a.value),2), a.year, a.period
FROM dbo.annual_2016 a

LEFT JOIN dbo.series s ON s.series_id = a.series_id
LEFT JOIN dbo.datatype d ON d.data_type_code = s.data_type_code

WHERE d.data_type_code = 30

GROUP BY a.year, a.period
````

2. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">In which industries did production and nonsupervisory employees fare better?</span>

In the Annual 2016 data, the top average weekly earnings were found in the <u>Pipeline Transportation</u>**,** <u>Software Publishers</u> and **<u>Petroleum/Coal Products</u>** industries.

In the January 2017 data, the top average weekly earnings were found in the <u>Reinsurance Carriers</u>**,** <span style="color: var(--vscode-foreground);"><u style="font-weight: bold;">Petroleum/Coal Products</u> and&nbsp;</span> <span style="color: var(--vscode-foreground);"><b><u>Fossil Fuel Electric Power Generation</u></b> industries.</span>

````sql
USE LaborStatisticsDB
SELECT round(avg(j.value),2) AS average_earnings, j.year, j.period, i.industry_name
FROM dbo.january_2017 j

LEFT JOIN dbo.series s ON s.series_id = j.series_id
LEFT JOIN dbo.datatype d ON d.data_type_code = s.data_type_code
LEFT JOIN dbo.industry i ON i.industry_code = s.industry_code

WHERE d.data_type_code = 30

GROUP BY j.year, j.period, i.industry_name

UNION 

SELECT round(avg(a.value),2), a.year, a.period, i.industry_name
FROM dbo.annual_2016 a

LEFT JOIN dbo.series s ON s.series_id = a.series_id
LEFT JOIN dbo.datatype d ON d.data_type_code = s.data_type_code
LEFT JOIN dbo.industry i ON i.industry_code = s.industry_code

WHERE d.data_type_code = 30

GROUP BY a.year, a.period, i.industry_name

ORDER BY average_earnings desc
````
